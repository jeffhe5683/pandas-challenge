```python
import pandas as pd
from pathlib import Path

# read CSV file and merge
school_data_load = Path('schools_complete.csv')
students_complete_load = Path('students_complete.csv')

school_data = pd.read_csv(school_data_load)
students_complete = pd.read_csv(students_complete_load)

school_data_complete = pd.merge(students_complete,school_data, on=["school_name"], how="left")

school_data_complete
```

```python
#calculate the total 
total_school = school_data["school_name"].count()
total_student= students_complete["student_name"].count()
total_budget= school_data["budget"].sum()
average_math_scores =students_complete["maths_score"].mean()
average_reading_scores=students_complete["reading_score"].mean()
number_math_pass = (students_complete['maths_score'] >= 50).sum()
percentage_math_pass = number_math_pass/total_student*100
number_reading_pass = (students_complete['reading_score']>=50).sum()
percentage_reading_pass = number_reading_pass/total_student*100
number_both_pass=((students_complete['maths_score']>=50)&(students_complete['reading_score']>=50)).sum()
percentage_both_pass=number_both_pass/total_student*100
sum_table = pd.DataFrame({
    'total_school': [total_school],
    'total_student': [total_student],
    'total_budget': [total_budget],
    'average_math_scores': [average_math_scores],
    'average_reading_scores': [average_reading_scores],
    'percentage_math_pass': [percentage_math_pass],
    'percentage_reading_pass': [percentage_reading_pass],
    'percentage_both_pass': [percentage_both_pass]
})
#formating data
sum_table['total_school'] = sum_table['total_school'].map("{:,}".format)
sum_table['total_student'] = sum_table['total_student'].map("{:,.0f}".format)
sum_table['total_budget'] = sum_table['total_budget'].map("${:,.2f}".format)
sum_table['average_math_scores'] = sum_table['average_math_scores'].map("{:,}".format)
sum_table['average_reading_scores'] = sum_table['average_reading_scores'].map("{:,}".format)
sum_table['percentage_math_pass'] = sum_table['percentage_math_pass'].map("{:,}".format)
sum_table['percentage_reading_pass'] = sum_table['percentage_reading_pass'].map("{:,}".format)
sum_table['percentage_both_pass'] = sum_table['percentage_both_pass'].map("{:,}".format)
sum_table
```

```python
#calcuate by school
school_type_by_school = school_data_complete["type"].groupby(school_data_complete['school_name']).first()
average_math_scores_by_school = students_complete["maths_score"].groupby(students_complete['school_name']).mean()
average_reading_scores_by_school = students_complete["reading_score"].groupby(students_complete['school_name']).mean()
number_math_pass_by_school = (students_complete['maths_score'] >= 50).groupby(students_complete['school_name']).sum()
total_student_by_school = school_data['size'].groupby(school_data['school_name']).first()
total_budget_by_school = school_data['budget'].groupby(school_data['school_name']).first()
budget_per_student = total_budget_by_school / total_student_by_school
percentage_math_pass_by_school = number_math_pass_by_school / total_student_by_school * 100
number_reading_pass_by_school = (students_complete['reading_score'] >= 50).groupby(students_complete['school_name']).sum()
percentage_reading_pass_by_school = number_reading_pass_by_school / total_student_by_school * 100
number_both_pass_by_school = ((students_complete['reading_score'] >= 50) & (students_complete['maths_score'] >= 50)).groupby(students_complete['school_name']).sum()
percentage_both_pass_by_school = number_both_pass_by_school / total_student_by_school * 100

sum_table_by_school = pd.DataFrame({
    'school_type': school_type_by_school,
    'total_students': total_student_by_school,
    'total_budget': total_budget_by_school,
    'per_student_budget': budget_per_student,
    'average_math_scores': average_math_scores_by_school,
    'average_reading_scores': average_reading_scores_by_school,
    '%passing_math': percentage_math_pass_by_school,
    '%passing_reading': percentage_reading_pass_by_school,
    'overall_passing': percentage_both_pass_by_school
})


sum_table_by_school['total_budget'] = sum_table_by_school['total_budget'].map("${:,.2f}".format)
sum_table_by_school['average_math_scores'] = sum_table_by_school['average_math_scores'].map("{:.2f}".format)
sum_table_by_school['average_reading_scores'] = sum_table_by_school['average_reading_scores'].map("{:.2f}".format)
sum_table_by_school['%passing_math'] = sum_table_by_school['%passing_math'].map("{:.2f}".format)
sum_table_by_school['%passing_reading'] = sum_table_by_school['%passing_reading'].map("{:.2f}".format)
sum_table_by_school['overall_passing'] = sum_table_by_school['overall_passing'].map("{:.2f}".format)

sum_table_by_school


```

```python
#select top 5 school by overall
top_performing_schools_by_overall = sum_table_by_school.sort_values('overall_passing', ascending=False).head(5)
top_performing_schools_by_overall


```

```python

#add column of bins
spending_bins = [0, 585, 630, 645, 680]
spending_labels = ['< $585', '$585 - $630', '$630 - $645', '$645 - $680']


sum_table_by_school["Spending Ranges (Per Student)"] = pd.cut(sum_table_by_school['per_student_budget'], bins=spending_bins, labels=spending_labels)

sum_table_by_school






```

```python

#calculate by spending ranges



sum_table_by_school['average_math_scores'] = pd.to_numeric(sum_table_by_school['average_math_scores'], errors='coerce')
sum_table_by_school['average_reading_scores'] = pd.to_numeric(sum_table_by_school['average_reading_scores'], errors='coerce')
sum_table_by_school['%passing_math'] = pd.to_numeric(sum_table_by_school['%passing_math'], errors='coerce')
sum_table_by_school['%passing_reading'] = pd.to_numeric(sum_table_by_school['%passing_reading'], errors='coerce')

spending_math_scores = sum_table_by_school.groupby(['Spending Ranges (Per Student)'])['average_math_scores'].mean().round(2)
spending_reading_scores = sum_table_by_school.groupby(['Spending Ranges (Per Student)'])['average_reading_scores'].mean().round(2)
spending_passing_math = sum_table_by_school.groupby(['Spending Ranges (Per Student)'])['%passing_math'].mean().round(2)
spending_passing_reading = sum_table_by_school.groupby(['Spending Ranges (Per Student)'])['%passing_reading'].mean().round(2)


spending_summary = pd.DataFrame({
    'average_math_scores': spending_math_scores,
    'average_reading_scores': spending_reading_scores,
    '%passing_math': spending_passing_math,
    '%passing_reading': spending_passing_reading,})

spending_summary
```
